# 부산광역시 교통·복지 연계형 취약계층 무더위 쉼터 플랫폼

## 프롬프트

AI는 친절한 코딩 튜터라고 가정한다. 이 프로젝트는 '부산광역시_무더위 쉼터 현황' CSV 파일과 판다스(Pandas), 그라디오(Gradio) 라이브러리만을 사용하여 완성도 높은 웹사이트를 제작하는 것을 목표로 한다. 사용자가 부산의 '구' 이름을 입력하면 실시간 통계 그래프와 쉼터 목록 표를 보여주고, 별도의 안전 가이드 탭까지 제공하는 파이썬 코드를 작성한다.

### 요구사항

#### 1. CSV 데이터 처리 및 동적 필터링/통계 로직
*   **CSV 파일 로드**: 코랩에 업로드된 무더위 쉼터 CSV 파일을 판다스(pd.read_csv)로 읽어온다. (한글 깨짐 방지를 위해 `encoding='cp949'` 예외 처리를 포함한다.)
*   **구별 데이터 추출 함수**: 사용자가 구 이름을 입력(예: '연제구')하면, 주소 컬럼에서 판다스의 `.str.contains(입력값, na=False)`를 사용해 해당 구의 데이터만 추출하는 파이썬 함수를 제작한다.
*   **함수 반환 값**: 이 함수는 다음 두 가지 결과물을 리턴해야 한다.:
    *   검색된 구의 '쉼터명', '상세주소', '쉼터유형' 컬럼만 남긴 데이터프레임
    *   해당 구의 '쉼터유형'별 개수(통계)를 집계한 데이터프레임 (그라디오 막대그래프인 `gr.BarPlot`에 연동할 수 있는 구조여야 한다.)
*   **예외 처리**: 만약 사용자가 입력한 구의 검색 결과가 없다면, 안내 메시지를 띄우는 예외 처리를 포함한다.

#### 2. Gradio 웹 UI 화면 구성 (탭 레이아웃 및 추가 기능 반영)
*   **취약계층 친화적 디자인**: 어르신들과 취약계층이 사용하기 좋게 폰트가 크고 직관적인 인터페이스를 구현한다.
*   **탭 레이아웃**: `gr.Blocks()` 내부에 `gr.Tab()`을 사용해 총 2개의 탭으로 화면을 구성한다.

##### [첫 번째 탭: '쉼터 검색 및 현황']
*   **상단**: 큰 글씨의 제목과 '원하시는 구 이름을 입력하시면 대피처 목록과 구별 통계를 보여드립니다'라는 안내 문구를 배치한다.
*   **입력 영역**: 사용자가 구 이름을 적는 텍스트 상자(`gr.Textbox(label="부산광역시 '구' 입력 (예: 연제구, 해운대구))")`와 '검색하기' 버튼을 배치한다.
*   **출력 영역 (실시간 통계 및 표)**: 버튼을 누르면 검색 결과가 나타나는 공간이다. `gr.Row()`를 사용해 좌우로 분할한다.
    *   **좌측**: 해당 구에 어떤 유형의 쉼터가 많은지 시각적으로 보여주는 그라디오 막대그래프 컴포넌트(`gr.BarPlot`)를 배치하여 시각적 완성도를 높인다. (X축: 쉼터유형, Y축: 개수)
    *   **우측**: 검색된 총 쉼터 개수를 알려주는 텍스트 상자와, 쉼터 이름 및 주소가 정렬되어 나오는 그라디오 데이터프레임 컴포넌트(`gr.Dataframe`)를 배치한다. 사용자가 검색할 때마다 그래프와 표가 실시간으로 동시에 갱신되어야 한다.

##### [두 번째 탭: '폭염 대처 국민행동요령']
*   **정보 페이지**: 이 탭은 검색창 없이 행정안전부 국민행동요령 정보를 큼직한 글씨로 깔끔하게 정리해 둔 정보 페이지이다.
*   **내용**: `gr.Markdown()` 등을 활용해 다음 내용을 어르신들이 읽기 편하게 가독성 높은 텍스트 구조로 보기 좋게 작성한다:
    *   1. 폭염 예보 시 준비사항 (물 자주 마시기, 햇볕 차단)
    *   2. 폭염 발생 시 행동요령 (낮 시간 야외활동 자제, 휴식 취하기)
    *   3. 온열질환(어지러움, 두통) 발생 시 응급처치법 (119 신고, 시원한 곳으로 이동)

*   **최종 스크립트**: 모든 판다스 데이터 가공, 그라디오 탭 레이아웃, 실시간 그래프 업데이트 로직이 에러 없이 완벽하게 유기적으로 돌아가는 단일 파이썬 스크립트 코드를 완성한다.
*   **배포**: 코드 맨 밑에 외부 테스트용 Public URL이 열리도록 `demo.launch(share=True)`를 붙인다.
상세한 한글 주석을 단다,

```
import pandas as pd
import gradio as gr

# 1. CSV 데이터 로드 및 주소 필터링 파트

# CSV 파일 로드 (encoding='cp949' 예외 처리 포함)
try:
    df_shelters_raw = pd.read_csv('부산광역시_무더위 쉼터_20251119 (1).csv', encoding='cp949')
except UnicodeDecodeError:
    print("cp949 인코딩으로 파일을 읽는 데 실패했습니다. utf-8 인코딩을 시도합니다.")
    df_shelters_raw = pd.read_csv('부산광역시_무더위 쉼터_20251119 (1).csv', encoding='utf-8')

# 필요한 컬럼 확인 및 기본 데이터프레임 준비
# '쉼터명', '상세주소'는 필수. '쉼터유형'은 파생해야 함.
required_cols_present = ['쉼터명', '상세주소']
missing_cols = [col for col in required_cols_present if col not in df_shelters_raw.columns]
if missing_cols:
    raise ValueError(f"CSV 파일에 다음 필수 컬럼이 없습니다: {', '.join(missing_cols)}")

df_base = df_shelters_raw[required_cols_present].copy()

# '쉼터유형' 컬럼 생성 함수 (쉼터명 또는 상세주소에서 키워드 추출)
def infer_shelter_type(name, address):
    name_lower = str(name).lower()
    address_lower = str(address).lower()
    if '경로당' in name_lower or '경로당' in address_lower:
        return '경로당'
    elif '행정복지센터' in name_lower or '행정복지센터' in address_lower:
        return '행정복지센터'
    elif '주민센터' in name_lower or '주민센터' in address_lower:
        return '주민센터'
    elif '복지관' in name_lower or '복지관' in address_lower:
        return '복지관'
    elif '마을회관' in name_lower or '마을회관' in address_lower:
        return '마을회관'
    elif '원로의집' in name_lower or '원로원' in name_lower:
        return '원로의집'
    elif '회관' in name_lower:
        return '회관'
    elif '도서관' in name_lower:
        return '도서관'
    elif '문화의집' in name_lower:
        return '문화의집'
    elif '센터' in name_lower: # Catch-all for other centers
        return '기타 센터'
    return '기타'

df_base['쉼터유형'] = df_base.apply(lambda row: infer_shelter_type(row['쉼터명'], row['상세주소']), axis=1)

# '구' 이름으로 쉼터 필터링 함수
def filter_shelters_by_gu(gu_name):
    if not gu_name or not gu_name.strip():
        empty_df = pd.DataFrame(columns=['쉼터명', '상세주소', '쉼터유형'])
        empty_stats_df = pd.DataFrame(columns=['쉼터유형', '개수'])
        return "구 이름을 입력해주세요.", empty_df, empty_stats_df

    # '상세주소' 컬럼에서 구 이름이 포함된 행 필터링
    filtered_df = df_base[df_base['상세주소'].str.contains(f"부산광역시 {gu_name.strip()}", na=False, regex=False)].copy()

    if filtered_df.empty:
        empty_df = pd.DataFrame(columns=['쉼터명', '상세주소', '쉼터유형'])
        empty_stats_df = pd.DataFrame(columns=['쉼터유형', '개수'])
        return f"검색된 쉼터가 없습니다. '{gu_name.strip()}' 구 이름을 다시 확인해 주세요.", empty_df, empty_stats_df
    else:
        # 표시용 데이터프레임
        display_df = filtered_df[['쉼터명', '상세주소', '쉼터유형']]

        # 통계용 데이터프레임 (BarPlot용)
        stats_df = filtered_df['쉼터유형'].value_counts().reset_index()
        stats_df.columns = ['쉼터유형', '개수'] # 컬럼명 지정

        message = f"'{gu_name.strip()}' 구에서 {len(display_df)}개의 쉼터를 찾았습니다."
        return message, display_df, stats_df

# 2. Gradio 웹 UI 화면 구성 파트
with gr.Blocks(title="우리 동네 무더위 쉼터 찾기 (부산광역시)") as demo:
    gr.Markdown(
        """
        <h1 style="font-family: 'Malgun Gothic', '맑은 고딕', sans-serif; text-align: center; font-size: 2.5em; color: #1e88e5;">
            우리 동네 무더위 쉼터 찾기 (부산광역시)
        </h1>
        <p style="font-family: 'Malgun Gothic', '맑은 고딕', sans-serif; text-align: center; font-size: 1.1em; color: #424242;">
            어르신들을 위한 부산 지역의 무더위 쉼터 정보와 폭염 대처 요령을 알려드립니다.
        </p>
        """
    )

    with gr.Tabs():
        with gr.Tab('쉼터 검색 및 현황'):
            gr.Markdown(
                """
                <h2 style="font-family: 'Malgun Gothic', '맑은 고딕', sans-serif; text-align: center; font-size: 1.8em; color: #333;">
                    원하시는 구 이름을 입력하시면 대피처 목록과 구별 통계를 보여드립니다.
                </h2>
                """
            )
            with gr.Row():
                gu_input = gr.Textbox(
                    label="부산광역시 '구' 입력 (예: 연제구, 부산진구)",
                    placeholder="예: 연제구",
                    scale=3,
                    elem_id="gu_input_textbox"
                )
                search_button = gr.Button("검색하기", scale=1)

            # 결과 메시지 및 데이터프레임 출력 영역
            result_message = gr.Markdown("", elem_id="result_message_markdown")

            with gr.Row():
                # 좌측: 통계 그래프
                with gr.Column(scale=2): # Changed scale from 1 to 2
                    shelter_type_barplot = gr.BarPlot(
                        x="쉼터유형",
                        y="개수",
                        title="쉼터 유형별 개수",
                        label="쉼터 유형별 개수",
                        x_title="쉼터 유형",
                        y_title="개수",
                        height=300,
                        elem_id="shelter_type_barplot"
                    )
                # 우측: 쉼터 목록
                with gr.Column(scale=1): # Changed scale from 3 to 1
                    shelter_output_df = gr.Dataframe(
                        headers=["쉼터명", "상세주소", "쉼터유형"],
                        wrap=True,
                        interactive=False, # Read-only
                        label="검색 결과",
                        elem_id="shelter_output_dataframe"
                    )

            # 이벤트 연동: 버튼 클릭 시 필터링 함수 실행
            search_button.click(
                fn=filter_shelters_by_gu,
                inputs=gu_input,
                outputs=[result_message, shelter_output_df, shelter_type_barplot]
            )

        with gr.Tab('폭염 대처 국민행동요령'):
            gr.Markdown(
                """
                <h2 style="font-family: 'Malgun Gothic', '맑은 고딕', sans-serif; text-align: center; font-size: 1.8em; color: #333;">
                    폭염 대처 국민행동요령
                </h2>
                <h3 style="font-family: 'Malgun Gothic', '맑은 고딕', sans-serif; color: #e65100;">1. 폭염 예보 시 준비사항</h3>
                <ul style="font-family: 'Malgun Gothic', '맑은 고딕', sans-serif; font-size: 1.1em; line-height: 1.6;">
                    <li>물 자주 마시기: 갈증을 느끼지 않아도 물을 자주 마시고, 땀을 많이 흘렸을 때는 스포츠 음료나 이온 음료를 마십니다.</li>
                    <li>햇볕 차단: 양산, 모자, 얇고 밝은 색 옷을 착용하여 햇볕 노출을 최소화합니다.</li>
                    <li>전력 사용량 확인: 에어컨, 선풍기 사용량이 늘어날 수 있으므로 전력 사용량을 확인하고 절약합니다.</li>
                    <li>외출 자제: 가장 더운 시간대(오전 12시~오후 5시)에는 외출을 자제하고, 특히 어르신이나 영유아는 더욱 주의합니다.</li>
                </ul>

                <h3 style="font-family: 'Malgun Gothic', '맑은 고딕', sans-serif; color: #e65100;">2. 폭염 발생 시 행동요령</h3>
                <ul style="font-family: 'Malgun Gothic', '맑은 고딕', sans-serif; font-size: 1.1em; line-height: 1.6;">
                    <li>낮 시간 야외활동 자제: 한낮에는 논밭 작업이나 야외 활동을 피하고, 불가피하게 야외 활동을 할 경우 챙이 넓은 모자, 밝은 색의 헐렁한 옷을 착용합니다.</li>
                    <li>휴식 취하기: 작업 중에는 규칙적으로 그늘에서 휴식을 취하고, 충분한 수분을 섭취합니다.</li>
                    <li>실내 시원하게 유지: 에어컨이나 선풍기를 사용하여 실내 온도를 시원하게 유지하고, 창문과 문을 닫아 뜨거운 공기가 들어오는 것을 막습니다.</li>
                    <li>무더위 쉼터 이용: 가까운 무더위 쉼터(경로당, 주민센터 등)를 방문하여 더위를 식힙니다.</li>
                </ul>

                <h3 style="font-family: 'Malgun Gothic', '맑은 고딕', sans-serif; color: #e65100;">3. 온열질환 발생 시 응급처치법</h3>
                <ul style="font-family: 'Malgun Gothic', '맑은 고딕', sans-serif; font-size: 1.1em; line-height: 1.6;">
                    <li>119 신고: 의식이 없으면 즉시 119에 신고하고, 구급차가 올 때까지 시원한 곳으로 옮겨 눕힙니다.</li>
                    <li>시원한 곳으로 이동: 환자를 시원한 그늘이나 에어컨이 있는 실내로 옮기고, 옷을 느슨하게 풀어줍니다.</li>
                    <li>몸 식히기: 물수건이나 얼음주머니로 몸을 닦거나 부채질하여 체온을 식힙니다.</li>
                    <li>수분 섭취: 의식이 있는 환자에게는 시원한 물이나 이온 음료를 천천히 마시게 합니다.</li>
                </ul>
                <p style="font-family: 'Malgun Gothic', '맑은 고딕', sans-serif; text-align: right; font-size: 0.9em; color: #757575; margin-top: 20px;">
                    출처: 행정안전부 국민재난안전포털
                </p>
                """
            )

# Gradio 앱 실행 (Public URL 제공)
demo.launch(share=True)
```
